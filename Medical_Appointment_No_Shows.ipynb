```python
import pandas as pd
```

```python
df=pd.read_csv('medical_appointment_no_shows.csv')
df
```

```python
df.isnull().sum()
```

##### No null values present in the dataset 

```python
df=df.drop_duplicates()
```

```python
df
```

##### The number of rows remained the same , that means there were no duplicates also in the dataset 

```python
df.columns=['Patient_ID','Appointment_ID','Gender','Scheduled_Day','Appointment_Day','Age','Neighbourhood','Scholarship','Hypertension','Diabetes','Alcoholism','Handicapped','SMS_Recieved','No-Show']
df
```

##### The column names has been standardised and the minor spelling mistakes (Hipertension) were fixed as well 

```python
df['Scheduled_Day']=df['Scheduled_Day'].str.replace('T',' ',regex=False)
```

```python
df['Scheduled_Day']=df['Scheduled_Day'].str.replace('Z','',regex=False)
```

```python
df['Appointment_Day']=df['Appointment_Day'].str.replace('T','',regex=False)
df['Appointment_Day']=df['Appointment_Day'].str.replace('Z','',regex=False)
```

```python
df
```

```python
df['Scheduled_Day']=df['Scheduled_Day'].str[:10] + ' ' + df['Scheduled_Day'].str[10:]
df
```

```python
df['Appointment_Day']=df['Appointment_Day'].str[:10] + ' ' + df['Appointment_Day'].str[10:]
df
```

```python
df['Scheduled_Day']=pd.to_datetime(df['Scheduled_Day'])
df['Appointment_Day']=pd.to_datetime(df['Appointment_Day'])
```

##### The Scheduled_Day and Appointment_Day columns are now in date-time formats 

```python
df['Scheduled_Date']=df['Scheduled_Day'].dt.date
df['Scheduled_Time']=df['Scheduled_Day'].dt.time
df['Appointment_Date']=df['Appointment_Day'].dt.date
df['Appointment_Time']=df['Appointment_Day'].dt.time
```

```python
df
```

```python
df['Scheduled_Date']=pd.to_datetime(df['Scheduled_Date'])
df['Scheduled_Month']=df['Scheduled_Date'].dt.month_name()
df['Scheduled_Day']=df['Scheduled_Date'].dt.day
```

```python
df
```

```python
df['Appointment_Date']=pd.to_datetime(df['Appointment_Date'])
df['Appointment_Month']=df['Appointment_Date'].dt.month_name()
df['Appointment_Day']=df['Appointment_Date'].dt.day
```

```python
df
```
